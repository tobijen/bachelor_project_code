This jupyter notebook is used for cleaning the fulltext. More precisely is used for removing citations and the special characters "\n" from the fulltexts.

```python
# import packages needed
import pandas as pd
import json
import re
```

```python
csv_file = 'path_to_dataset/complete_allsides_data.csv'

# Read CSV file using pandas
data_df = pd.read_csv(csv_file)
data_df
```

## Remove Special Characters from text

```python
# function to remove the special character "\n"
def remove_new_lines(text):
    return re.sub(r'\n', '', text)
```

```python
# apply the function to the fulltext column and write the results in the column full_text_cleaned
data_df['full_text_cleaned'] = data_df['full_text']
data_df['full_text_cleaned'] = data_df['full_text_cleaned'].fillna('No text found')
data_df['full_text_cleaned'] = data_df['full_text_cleaned'].apply(remove_new_lines)
data_df['full_text_cleaned'][7]
```

## Remove Quotations from full_text

```python
# function to remove the citations
def remove_quotes(text):
    return re.sub(r'"([^"]*)"', '', text)
```

```python
# apply the function to the newly created column full_text_cleaned
data_df['full_text_cleaned'] = data_df['full_text_cleaned'].apply(remove_quotes)
data_df['full_text_cleaned'][7]
```

```python
# write the cleaned data into a new csv_file
data_df.to_csv("path_to_store_the_cleaned_dataset/complete_allsides_data_cleaned.csv")
```
