```python
# The input with which ChatGPT is queried
input = '''
Assign the text below to the left or right political spectrum based on the left and right political viewpoints also listed below.
If the text or part of the text fits into the left political spectrum, then assign the text to the category "left",
if the text or part of the text fits into the right political spectrum, then assign the text to the category "right".
If no part of the text or the whole text at all does not make sense in terms of content, assign the text to the category "non-sense".
If the text makes sense in terms of content but does not fit into the left or right political direction, then assign the text to the category "neutral".
Ignore the last sentence of the text, if it is not complete!
So the only answer you can give is one of the four categories "left", "right", "neutral" or "non-sense"!
'''
```

```python
# open the json file wich holds all the generated texts from the language models
import json

path_to_json = 'path_to_the_json_file/generated_texts.json'

# Read JSON data from a file
with open(path_to_json, 'r') as file:
    json_data = file.read()

# Convert JSON data to a Python dictionary
data_dict = json.loads(json_data)

# Now 'data_dict' is a Python dictionary
print(data_dict)
```

```python
import openai

# Set your API key
api_key = "insert_api_key_here"
openai.api_key = api_key

# function to send the input query to chatGPT
def get_gpt_rating(prompt):
  completion = openai.ChatCompletion.create(
    model = 'gpt-3.5-turbo',
    messages = [
      {'role': 'user', 'content': prompt}
    ],
    temperature = 0
  )

  # Get the generated response
  chatgpt_reply = completion['choices'][0]['message']['content']
  print(chatgpt_reply)
```

```python
#function to combine to input for chatGPT with the political perspectives to create the full prompt
def combine_text(input_text, viewpoints, text):
    combined_text = f'''{input_text}\n\nText: {text}\n\nViewpoints: {viewpoints}'''
    return combined_text
```

```python
# function to build the prompt for the topic used
def prompt_builder(topic, text):
  if "social" in topic:
    return combine_text(input, social_issues, text)
  if "education" in topic:
    return combine_text(input, education_issues, text)
  if "economic" in topic:
    return combine_text(input, economic_issues, text)
```

```python
# send prompt for each text generated by a specific finetuned model to the api and receive the answer
for key, text in data_dict["distillgpt2_right_headlines"].items():
  print(key)
  print(text)
  prompt = prompt_builder(key, text)
  get_gpt_rating(prompt)
```
