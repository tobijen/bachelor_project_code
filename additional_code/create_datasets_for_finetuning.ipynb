The purpose of this note boot is to create the different datasets, needed for finetuning the language models. The database for the different datasets is the complete_allsides_data_cleaned.csv. Six additional data sets are created from the data set. Two for each political spectrum. One of the two contains to the headings without null_values and the other contains the fulltext and cleaned fulltext without null_values.

```python
# import packages
import pandas as pd
```

```python
csv_file = 'path_to_dataset/complete_allsides_data_cleaned.csv'

# Read CSV file using pandas
data_df = pd.read_csv(csv_file)
```

# Create left data


### heading

```python
# create dataset for left headings without null values
data_df_left = data_df[data_df["bias_rating"]=="left"]
data_df_left_heading = data_df_left[data_df_left[['heading']].notnull().all(1)]
len(data_df_left_heading)

data_df_left_heading.to_csv("path_to_store_dataset/allsides_data_left_heading.csv")
```

### fulltext and cleaned fulltext

```python
# create dataset for left fulltext and cleaned fulltext without null values
data_df_left = data_df[data_df["bias_rating"]=="left"]
data_df_left_fulltext = data_df_left[data_df_left[['full_text']].notnull().all(1)]
len(data_df_left_fulltext)

data_df_left_fulltext.to_csv("path_to_store_dataset/allsides_data_left_fulltext.csv")
```

### fulltext and cleaned fulltext

```python
# create dataset for center fulltext and cleaned fulltext without null values
data_df_center = data_df[data_df["bias_rating"]=="center"]
data_df_center_fulltext = data_df_center[data_df_center[['full_text']].notnull().all(1)]
len(data_df_center_fulltext)

data_df_center_fulltext.to_csv("path_to_store_dataset/allsides_data_center_fulltext.csv")
```
