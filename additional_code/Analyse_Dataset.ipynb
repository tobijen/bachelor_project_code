```python
# read datasets
csv_file = 'path_to_dataset/complete_allsides_data.csv'
csv_file_cleaned = 'path_to_dataset/complete_allsides_data_cleaned.csv'

data_df = pd.read_csv(csv_file_cleaned)
data_df
```

```python
# Assuming your dataframe is called 'df' and 'fulltext' is the column containing the text
def word_count(text):
    if pd.isnull(text):
        return 0
    else:
        words = text.split()
        return len(words)
```

```python
# check if data contains '\n'
data_df[data_df['full_text_cleaned'].str.contains('\n', na=False)].shape[0]
```

```python
# check if data contains quotes
pattern = r'"([^"]*)"'
data_df[data_df['full_text_cleaned'].str.contains(pattern, regex=True, na=False)].shape[0]
```

```python
# get the average appearance of quotes in full texts

mask = data_df['full_text'].apply(lambda x: isinstance(x, str))

# Find all occurrences of the regex pattern in the filtered 'full_text' column
matches = data_df.loc[mask, 'full_text'].str.findall(pattern)

# Count the number of matches per row
match_counts = [len(match) for match in matches]

# Calculate the average appearance of matches per row
average_appearance = sum(match_counts) / len(data_df[mask])

average_appearance
```

```python
# check null values
data_df.isna().sum().sum()
```

```python
len(data_df)
```

```python
# get unique count of news article sources
data_df['source'].nunique()
```

```python
# get avg word count
data_df['word_count'] = data_df['full_text_cleaned'].apply(word_count)
data_df['word_count'].mean()
```
